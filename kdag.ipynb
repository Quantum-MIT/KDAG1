```python
def create_dataset(data, k):
    dataX, dataY = [],[]
    for i in range(data.shape[0] - k):
        x = data[i:i + k, :]
        y = data[i + k, :]
        dataX.append(x)
        dataY.append(y)
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 12
trainX, trainY = create_dataset(train_arr, look_back)
testX, testY = create_dataset(test_arr, look_back)
```
